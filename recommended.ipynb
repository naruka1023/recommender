```python
import pandas as pd
import numpy as np

import sqlite3 as sql

con = sql.connect("eComDB.db")
cur = con.cursor()
cur.execute("SELECT * FROM userData")
rows = cur.fetchall()
df = pd.DataFrame(rows, columns=['transactionID', 'userID', 'ItemID'])
df
```

```python
cur.execute("SELECT id, title FROM items")
rows = cur.fetchall()
item = pd.DataFrame(rows, columns=['ItemID', 'foodTitle'])
item.shape
```

```python
df_merge = pd.merge(df, item, how='right')
df_merge['flag'] = 1
df_merge1 = df_merge[df_merge['userID']!=1.0]
df_merge2 = df_merge[df_merge['userID']==1.0]
#df_merge = df_merge.sort_values(by='ItemID', ascending=True)
df_merge1['flag'] = 0
df_merge = df_merge2.append(df_merge1)
df_merge['userID'] = 1.0
df_merge
```

```python
userToFood = df_merge.pivot_table(index='userID',columns='foodTitle',values='flag')
userToFood.head(2)
```

```python
matrix_corr = userToFood.corr(method='pearson',min_periods=100)
matrix_corr.head()
```
