```python
import pandas as pd
import numpy as np

import sqlite3 as sql

con = sql.connect("eComDB.db")
cur = con.cursor()
cur.execute("SELECT * FROM userData")
rows = cur.fetchall()
df = pd.DataFrame(rows, columns=['transactionID', 'userID', 'ItemID'])
df.head()
```

```python
cur.execute("SELECT id, title FROM items")
rows = cur.fetchall()
item = pd.DataFrame(rows, columns=['ItemID', 'foodTitle'])
item.head()
```

```python
df_merge = pd.merge(df, item)
df_merge['flag'] = 1
df_merge.head()
```

```python
userToFood = df_merge.pivot_table(index='userID',columns='foodTitle',values='flag')
userToFood.head(2)
```

```python
matrix_corr = userToFood.corr(method='pearson',min_periods=100)
matrix_corr.head()
```
